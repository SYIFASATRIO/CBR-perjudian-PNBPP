```python
!pip install pypdf -q

import os
from pypdf import PdfReader

folder = "/content/drive/MyDrive/PN BPP"

cases = []

for file in os.listdir(folder):
    if file.endswith(".pdf"):
        path = os.path.join(folder, file)

        try:
            reader = PdfReader(path)

            text = ""
            for page in reader.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text + " "

            cases.append({
                "filename": file,
                "text": text
            })

            print("Berhasil:", file)

        except Exception as e:
            print("Error:", file, e)

print("Jumlah dokumen:", len(cases))
```

```python
import pandas as pd

df = pd.DataFrame(cases)

df.head()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text

df["clean_text"] = df["text"].apply(clean_text)
```

```python
import pandas as pd

df = pd.DataFrame(cases)

print(df.shape)
df.head()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\n+', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    return text.strip()

df["clean_text"] = df["text"].apply(clean_text)
```

```python
def extract_case_number(filename):
    match = re.search(r'putusan_(.*?)_pn_bpp', filename)
    if match:
        return match.group(1)
    return None

df["case_number"] = df["filename"].apply(extract_case_number)
```

```python
df["word_count"] = df["clean_text"].apply(
    lambda x: len(x.split())
)
```

```python
df.to_csv(
    "/content/drive/MyDrive/PN BPP/cases.csv",
    index=False
)

print("CSV berhasil disimpan")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words=None,
    max_features=5000
)

tfidf_matrix = vectorizer.fit_transform(df["clean_text"])

print("Shape TF-IDF:", tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(tfidf_matrix)

print(similarity_matrix.shape)
```

```python
def retrieve(query, k=5):

    query_clean = clean_text(query)

    query_vector = vectorizer.transform([query_clean])

    similarities = cosine_similarity(
        query_vector,
        tfidf_matrix
    ).flatten()

    top_indices = similarities.argsort()[-k:][::-1]

    results = []

    for idx in top_indices:
        results.append({
            "case_number": df.iloc[idx]["case_number"],
            "filename": df.iloc[idx]["filename"],
            "similarity": similarities[idx]
        })

    return pd.DataFrame(results)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def retrieve(query, k=5):

    query = clean_text(query)

    query_vector = vectorizer.transform([query])

    similarities = cosine_similarity(
        query_vector,
        tfidf_matrix
    ).flatten()

    top_indices = similarities.argsort()[-k:][::-1]

    return pd.DataFrame({
        "case_number": df.iloc[top_indices]["case_number"].values,
        "filename": df.iloc[top_indices]["filename"].values,
        "similarity": similarities[top_indices]
    })
```

```python
import re

def extract_verdict(text):

    pattern = r"mengadili(.*?)(?:demikian diputuskan|$)"

    match = re.search(
        pattern,
        text.lower(),
        re.DOTALL
    )

    if match:
        return match.group(1)[:1500]

    return ""

df["verdict"] = df["text"].apply(extract_verdict)
```

```python
def predict_outcome(query):

    result = retrieve(query, k=5)

    best_case = result.iloc[0]

    idx = df[
        df["case_number"] ==
        best_case["case_number"]
    ].index[0]

    return df.loc[idx, "verdict"]
```

```python
import re

def extract_defendant(text):
    match = re.search(
        r'Nama lengkap\s*:\s*(.*?)(?:\n|Tempat lahir)',
        text,
        re.IGNORECASE | re.DOTALL
    )

    if match:
        return match.group(1).strip()

    return ""

df["terdakwa"] = df["text"].apply(extract_defendant)
```

```python
def extract_nomor_putusan(text):

    match = re.search(
        r'Nomor\s+(\d+/Pid\.B/\d+/PN\s+\w+)',
        text,
        re.IGNORECASE
    )

    if match:
        return match.group(1)

    return ""

df["nomor_putusan"] = df["text"].apply(
    extract_nomor_putusan
)
```

```python
def extract_pasal(text):

    text = text.lower()

    if "303 bis" in text:
        return "303 bis KUHP"

    if "pasal 303" in text:
        return "303 KUHP"

    return "Tidak ditemukan"

df["pasal"] = df["text"].apply(extract_pasal)
```
